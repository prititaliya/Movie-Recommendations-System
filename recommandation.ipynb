```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
df=pd.merge(movies,ratings,on='movieId')
```

```python
df.index=df['userId']
df.drop('userId',axis=1,inplace=True)
```

```python
df.drop(['genres','timestamp'],axis=1,inplace=True)
```

```python
# popular_movies = df.groupby('title')['rating'].agg(['mean', 'count']).sort_values(by='count', ascending=False).head(10)
temp = df.groupby('title').filter(lambda x: x['rating'].count() > 5000)
temp
```

```python
ndf=temp.pivot_table(index='userId',columns='title',values='rating')
# temp['title'].value_counts()
```

```python
ndf.dropna(thresh=10,axis=1,inplace=True)
ndf
```

```python
ndf.fillna(0,inplace=True)
# ndf.isna().sum()
```

```python
simmilarity=ndf.corr(method='pearson') #pearson correlation
# simmilarity
```

```python
def get_similar(movie_name,rating):
    similar_ratings = simmilarity[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    return similar_ratings
```

```python
get_similar("Transformers (2007)",4).head(10)
```
